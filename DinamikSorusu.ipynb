```python
import matplotlib.pyplot as plt
import numpy as np
from math import pi
```

```python
def vc(tetha):
    beta = np.arcsin((86/89)*np.sin(tetha))
    _vc = (933.96*np.sin(tetha+beta))/np.cos(beta)
    return _vc
```

```python
def vb(x):
    vb = 933.96
    return vb
```

```python
interval = [1,2*pi]
```

```python
xpoints = np.arange(*interval,0.01)
ypoints = np.array([vc(x) for x in xpoints])
```
